```python
import re
from textblob import Word
def preprocess(x):
    x = x.replace(",000,000", " m").replace(",000", " k").replace("′", "'").replace("’", "'")\
                           .replace("won't", " will not").replace("cannot", " can not").replace("can't", " can not")\
                           .replace("n't", " not").replace("what's", " what is").replace("it's", " it is")\
                           .replace("'ve", " have").replace("'m", " am").replace("'re", " are")\
                           .replace("he's", " he is").replace("she's", " she is").replace("'s", " own")\
                           .replace("%", " percent ").replace("₹", " rupee ").replace("$", " dollar ")\
                           .replace("€", " euro ").replace("'ll", " will").replace("how's"," how has").replace("y'all"," you all")\
                           .replace("o'clock"," of the clock").replace("ne'er"," never").replace("let's"," let us")\
                           .replace("finna"," fixing to").replace("gonna"," going to").replace("gimme"," give me").replace("gotta"," got to").replace("'d"," would")\
                           .replace("daresn't"," dare not").replace("dasn't"," dare not").replace("e'er"," ever").replace("everyone's"," everyone is")\
                           .replace("'cause'"," because")
    
    x = re.sub(r"([0-9]+)000000", r"\1m", x)
    x = re.sub(r"([0-9]+)000", r"\1k", x)
    x=re.sub(r'((www\.[^\s]+)|(https?://[^\s]+))',' ',x)
    x=re.sub(r"\\s*\\b(?=\\w*(\\w)\\1{2,})\\w*\\b",' ',x)
    x=re.sub(r'<.*?>',' ',x)
    x=re.sub('[^a-zA-Z]',' ',x)

    return x


# Apply the clean_text function to the 'reviewText' column
df['cleaned_reviewText'] = df['reviewText'].apply(preprocess)
```

```python
df1 = df.iloc[:,[5,-1]]
df1.head()
```

```python
def score_classify(x):
    if x>3:
        return 'Positive'
    elif x<3:
        return 'Negative'
    else:
        return 'Neutral'
df1['Overall_Sentiment']=df1.apply(lambda x: score_classify(x['overall']),axis=1)
df1.head()
```

```python
df1.Overall_Sentiment.value_counts()
```

```python


sentiment_score_list = []
sentiment_label_list = []

for i in df1['cleaned_reviewText'].values.tolist():
    sentiment_text=TextBlob(i)
    sentiment_score = sentiment_text.sentiment.polarity
    #print(sentiment_score)

    if sentiment_score > 0:
        sentiment_score_list.append(sentiment_score)
        sentiment_label_list.append('Positive')
    elif sentiment_score < 0:
        sentiment_score_list.append(sentiment_score)
        sentiment_label_list.append('Negative')
    else:
        sentiment_score_list.append(sentiment_score)
        sentiment_label_list.append('Neutral')
    
df1['cleaned_reviewText'] = sentiment_label_list
df1['sentiment score'] = sentiment_score_list

display(df1.head(10))
```

```python
before=df1.shape[0]
df1 = df1[df1.Overall_Sentiment != 'Neutral']
df1 = df1[df1.cleaned_reviewText != 'Neutral']
df1.head(10)
after=df1.shape[0]
```

```python
def analyze_sentiments(cleaned_verified_reviews):
    analysis=TextBlob(cleaned_verified_reviews)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

Review_Senti=analyze_sentiments("The product is not that ")
Review_Senti
```

```python
def predict(user_review, user_score):
    user_score_senti = score_classify(user_score)
    user_review = user_review.lower()
    user_review = user_review.replace('[^\w\s]', '')
    user_review = preprocess(user_review)

    if user_review == '':
        print('Please Enter some review')
        return
    elif analyze_sentiments(user_review) == 'Neutral' or user_score_senti == 'Neutral':
        print('Review Submitted! Genuine')
    elif analyze_sentiments(user_review) == user_score_senti:
        print('Review Submitted! Genuine')
    elif analyze_sentiments(user_review) != user_score_senti:
        print('Review Submitted! Not Genuine. Please recheck your review.')

# Example usage
predict(user_review="The product is good", user_score=3)
```

```python
from joblib import dump, load
from textblob import TextBlob

# Save the sentiment analysis model
sentiment_model_filename = 'fake_review_detection_model.joblib'
dump(analyze_sentiments, sentiment_model_filename)

# Load the sentiment analysis model in another Python file
loaded_sentiment_model = load(sentiment_model_filename)

# Example usage in another Python file
user_review = "The product is good"
user_score = 3
predicted_sentiment = loaded_sentiment_model(user_review)

# You can use the predicted sentiment in your application
print(f"The predicted sentiment for the review is: {predicted_sentiment}")
```
